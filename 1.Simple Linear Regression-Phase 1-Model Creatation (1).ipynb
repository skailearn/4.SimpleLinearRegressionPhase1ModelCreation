```python
#1. Pandas is a library and it contains the functions related to the database
import pandas as pd
```

```python
#2. Reading csv data and its name of the file is Salaray_Data.csv
#3. Assigning that to the dataset variable.
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#4. display the vales from the dataset 
dataset
```

```python
# 5. Dataset is fetched only with "YearExperice" column
# 6. Assigning to the independent field dataset
independent=dataset[["YearsExperience"]]#
#.7 Displaying the values from the independent field
independent
```

```python
# 8. dataset is fetched only with "Salary" column
# 9. Assigning that value to the dependent field
dependent=dataset[["Salary"]]
#10. Displaying the values from the dependent field
dependent
```

```python
#11. importing the train_test_split class from sklearn.model.selection
from sklearn.model_selection import train_test_split
#12. 70 % and 30 % of the dataset is splited to train dataset and test dataset.
#13. train_test_split method is called with indepnedent, dependent,test size and random state arguments
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
#14. Checking split is done on train and test data set.
print(" X_train rows and column count are :",X_train.shape)
print(" y_train rows and column count are :", y_train.shape)
print(" X_test rows and column count are :",X_test.shape)
print(" y_test rows and column count are :", y_test.shape)
```

```python
#15. import LinearRegression from the sklearn.linear_model 
from sklearn.linear_model import LinearRegression
#16. LinearRegression assigned to the field name regressor
regressor=LinearRegression()
#17. Creating a model by calling fit method by passing X_train and y_train. so that it learns from the data set and create a model
regressor.fit(X_train,y_train)
```

```python
#18. As y = weight(x) + bias is the model equation . we are trying to display the value of the weight.
weight=regressor.coef_
weight
```

```python
#19. As y = weight(x) + bias is the model equation . we are trying to display the value of the bias.
bais=regressor.intercept_
bais
```

```python
#20. As y = 9360.26128619(x) + 26777.3913412 is the model equation . we are trying to predict the salary values for the X_test data set..
y_pred=regressor.predict(X_test)
```

```python
#21. Predictded values for the X_test data set
y_pred
```

```python
#22. import the r2_score from sklearn.metrics
from sklearn.metrics import r2_score
#23. Comparing the predicted Salary value and Expected value. so that we can measure the performance of the model.
#24. Passing the y_test and y_pred to the r2_score method and result is stored it in r_score field.
r_score=r2_score(y_test,y_pred)
```

```python
#25. displaying the value of the r_score value.
#26. if the score value is near to 1 then model performance is good.
r_score
```

```python
#27. importing the pickle to save the model
import pickle
#28. Assigning the file name to filename field.
filename="finalized_model_linear.sav"
```

```python
#29. Saving the model to that filename
pickle.dump(regressor,open(filename,'wb'))
```

```python
#30. loading the model and assignbed to the field loaded_model.
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
#31. Using that model predicting the Salary for the 13 years of Experienced person.
result=loaded_model.predict([[13]])
```
